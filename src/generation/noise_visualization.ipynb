```python
import numpy as np
import matplotlib.pyplot as plt

from noise_3d import fractal_perlin_noise_3d
```

```python
def visualize_noise_slice(size=(256, 256), scale=0.1, z_level=0.0, octaves=4, persistence=0.5, lacunarity=2.0, seed=0):
    width, height = size
    image = np.zeros(size)

    print(f"Generation {width}x{height}...")

    for i in range(width):
        for j in range(height):
            val = fractal_perlin_noise_3d(i, j, z_level, scale, octaves, persistence, lacunarity, seed)
            image[i, j] = (val + 1.0) / 2.0

    plt.figure(figsize=(8, 8))
    plt.imshow(image, cmap='magma', origin='lower')
    plt.colorbar(label='Amplitude')
    plt.title(f"3D Perlin Noise Scale={scale}")
    plt.show()
```

```python
visualize_noise_slice(size=(256, 256), scale=0.02, octaves=5, persistence=0.4, lacunarity=2, seed=0)
```
